```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("Zomato data .csv")
```

```python
dataframe
```

#convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()  # check the summary and null value in dataset
```

```python
# Type of Resturant
dataframe.head()
```

```python
# Define your colors
palette = sns.color_palette(["red", "green", "yellow", "pink"])

# Use the same variable name in countplot
sns.countplot(x=dataframe['listed_in(type)'], palette=palette,hue_order=True)
plt.xlabel("Type of resturant")
```

```python
# conclusion.........majority of the resturant falls in dinning category
```

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result.index, result['votes'], c="green", marker="o")
plt.xlabel("Type of restaurant", color="red", fontsize=20)
plt.ylabel("Votes", color="red", fontsize=20)
```

```python
# Conclusion....Dinning resturants has recieved maximum votes
```

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Rating distribution")
plt.show()
```

```python
# conclusion....The majority of the resturant getting 3.5 to 4.5 rating
```

```python
# Average order spending by couples
```

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
# conclusion .........the majority of the couples prefer resturants with an approximate cost of 300 rs
```

```python
# to find online mode or offline mode
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data=dataframe)
```

```python
 #Conclusion.....................offline order recieve lower rating
```

```python
pivot_table = dataframe.pivot_table(
    index='listed_in(type)',
    columns='online_order',
    aggfunc='size',
    fill_value=8
)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

```python
# Dinning resturant primarily accept offline orders, where as cafes primarily receive online orders
```
